```python
#Training the Markov Probabilistic Model
import random
from collections import defaultdict

class MarkovTextGenerator:
    def __init__(self):
        self.model = defaultdict(list)

    def train(self, text):
        words = text.lower().split()
        for i in range(len(words) - 1):
            self.model[words[i]].append(words[i + 1])

    def generate(self, start_word, length=30):
        start_word = start_word.lower()

        if start_word not in self.model:
            raise ValueError("Start word not found in training data")

        current_word = start_word
        generated = [current_word]

        for _ in range(length - 1):
            next_words = self.model[current_word]

            # Handle terminal words safely
            if not next_words:
                current_word = random.choice(list(self.model.keys()))
            else:
                current_word = random.choice(next_words)

            generated.append(current_word)

        return " ".join(generated)
```

```python
text = """
machine learning is fascinating
machine learning enables intelligent systems
learning from data is powerful
artificial intelligence is transforming technology
"""
```

```python
generator = MarkovTextGenerator()
generator.train(text)

output = generator.generate(start_word="machine", length=10)
print(output)
```
