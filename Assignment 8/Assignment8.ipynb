Text-to-Speech using Tacotron 2 and WaveNet



```python
!pip install pyttsx3
```

```python
import pyttsx3
```

```python
# Initialize TTS engine
engine = pyttsx3.init()
```

```python
# Configure voice properties
engine.setProperty('rate', 150)    # Speed
engine.setProperty('volume', 1.0)  # Volume (0.0 to 1.0)
```

```python
# Text to speech conversion 
text = "Namaste everyone! and a healthy good morning to all of you, This is a text to speech demo. I can convert any text to speech instantly. this was a great day and new experiences "
```

```python
# Generate speech and save to file
output_file = "output_pyttsx3.wav"
engine.save_to_file(text, output_file)
engine.runAndWait()
print(f"Speech generated and saved to {output_file}")
print("This method works offline and with any Python version!")
```

```python
# Play the audio in notebook (optional)
from IPython.display import Audio, display

display(Audio(output_file))
```

Method 2: Using Google Text-to-Speech (gTTS) - Simplest Method



```python
!pip install gTTS
```

```python
from gtts import gTTS
import os

# Text to convert to speech
text = "Welcome to text to speech conversion! This is a simple and easy method."

tts = gTTS(text=text, lang='en', slow=False)

output_file = "output_gtts.mp3"
tts.save(output_file)

print(f"Speech generated and saved to {output_file}")

# Play audio
from IPython.display import Audio, display
display(Audio(output_file))
```

Summary

**Two methods provided:**

1. **pyttsx3**: Works offline, compatible with ALL Python versions (including 3.12+)
2. **gTTS**: Simplest online method, fast and reliable

**Note:** Coqui TTS (Tacotron 2) requires Python lower than version 3.12, so we are using pyttsx3 instead which works with all Python versions.
